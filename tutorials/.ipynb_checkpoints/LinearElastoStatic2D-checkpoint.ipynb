# Solving linear elasticity ODE 

### Linear elasticity equation (case of small deformations)

```python
# small deformations (e.g. propagation of waves in metal) --> terms in ùõø/ùõøt are neglected (static)
```

$\left\{ 
\begin{array}
    &‚àí\nabla.\sigma=f& \mbox{in ‚Ñ¶}\\
    \sigma = \lambda tr(\epsilon)I + 2\mu\epsilon\\
    \epsilon = {1/2}(\nabla u + (\nabla u)^{T})\\
  \end{array}
  \right.$

### Variational form

```python
# Initialize spatial discretization of the PDE
```

Find u‚ààV such that: 

$\int\limits_\Omega \sigma(u):\epsilon(v) dx = \int\limits_\Omega f.v dx + \int\limits_{\delta\Omega} T.v ds$

### Principles of resolution with FEM 

The Finite Element Method (ex: FEniCS) provides a solution 'uh' that approximates the real solution 'u'.
The precision of uh depends on the number of basis functions (finite elements) considered and on their on their degree.

- u ~¬†uh=Œ£Œ±i.œÜi. œÜi are N basis functions used to discretize the spatial domain and solve the equation. œÜi are choosen to be orthogonal. Œ±i are the new N unknowns at each time step.
- v is choosen as a "test function" equal to œÜj. It enables to get N independant equations

The Finite Element Method uses the variational form to express the equation into a linear matrix problem: AU = L (linear variational problem in terms of 'u')

# Resolution with FEniCS (Finite Element Method library)

```python
import fenics
```

### Define domain

```python
mesh = fenics.RectangleMesh(fenics.Point(0., 0.), fenics.Point(2., 1.), 20, 20, "right/left")
#mesh = fenics.UnitSquareMesh(32, 32)
```

```python
fenics.plot(mesh)
```

### Define the space of basis functions to discretize the spatial domain

```python
# "Function Space": space of finite basis functions œÜi ("finite elements") generated in correlation with the domain. They will be used to approximate the real solution 'u'
```

```python
V = fenics.VectorFunctionSpace(mesh, "Lagrange", 1) #¬†family of functions --> Lagrange; degree --> 1 (linear polynome)
```

**N.B. Here, we use a VectorFunctionSpace and not a FunctionSpace of basis functions, since the unknown 'u' is the displacement vector.**

### Trial and test functions

```python
# "Function": composition of the finite basis functions in V
```

```python
u_trial = fenics.TrialFunction(V) #¬†trial function standing for 'uh' approximation -->  Œ£Œ±i*œÜi 
v_test = fenics.TestFunction(V) #¬†test function -->  œÜj
```

### Kinematics

$u = x - X$ (displacement)

$\epsilon = F - I$, with $\epsilon = \delta u / \delta x$ (strain)

```python
def epsilon(u):
    return fenics.grad(u) 
```

### Material

```python
#¬†Linear elasticity
```

$\sigma = \lambda tr(\epsilon)I + 2\mu\epsilon$

```python
# Parameters
E = fenics.Constant(1e5)
nu = fenics.Constant(0.3)

mu = E/2/(1+nu)
lmbda = E*nu/(1+nu)/(1-2*nu)
```

```python
#¬†Constitutive model
def sigma(u):
    return lmbda*fenics.tr(epsilon(u))*fenics.Identity(2) + 2.0*mu*epsilon(u)
```

### Boundaries

**N.B** 

**Defining and marking boundaries enables to**:
- **apply surface forces on specific parts of the boundary**
- **define Dirichlet or Neumann conditions on specific parts of the boundary**


**Defining and marking subdomains enables to**:
- **apply volume forces or material properties on specific parts of the domain**

https://fenicsproject.org/pub/tutorial/sphinx1/._ftut1005.html

```python
#¬†Used here to define Dirichlet boundary conditions
```

```python
#Mesh boundaries
boundaries = fenics.MeshFunction("size_t", mesh, mesh.topology().dim() - 1) 

#¬†Provide mark to all boundaries
boundaries.set_all(0)
```

#### Bottom fixed

```python
# Define Bottom boundary
class Bottom(fenics.SubDomain):
    def inside(self, x, on_boundary):
        tol = 1E-14
        return on_boundary and fenics.near(x[1], 0, tol)

bottom = Bottom()

#¬†mark Left boundary with 99
bottom.mark(boundaries, 99)
```

#### Left fixed

```python
# to define fixed left Dirichlet bc
```

```python
# Define Left boundary
class Left(fenics.SubDomain):
    def inside(self, x, on_boundary):
        tol = 1E-14
        return on_boundary and fenics.near(x[0], 0, tol)

left = Left()

#¬†mark Left boundary with 1
left.mark(boundaries, 1)
```

#### Right fixed

```python
# Define Right boundary
class Right(fenics.SubDomain):
    def inside(self, x, on_boundary):
        tol = 1E-14
        return on_boundary and fenics.near(x[0], 1, tol)

right = Right()

#¬†mark Left boundary with 1
right.mark(boundaries, 2)
```

```python
#¬†define "ds" integration Measure
ds = fenics.Measure("ds", domain=mesh, subdomain_data=boundaries)
```

### Variational form

$\int\limits_\Omega \sigma(u):\epsilon(v) dx = \int\limits_\Omega f.v dx + \int\limits_{\delta\Omega} T.v ds$

with:
- $a(u,v)$ = $\int\limits_\Omega \sigma(u):\epsilon(v) dx$ (bilinear form)

- $L(v)$ = $\int\limits_\Omega f.v dx + \int\limits_{\delta\Omega} T.v ds$ (linear form)

#### bilinear form a(u_trial,v_test)

```python
#¬†a(u_trial,v_test) is part of the variational form depending on the 'uh' solution
```

```python
a = fenics.inner(sigma(u_trial), epsilon(v_test)) * fenics.dx
```

#### linear form L(v_test) 

```python
# L(v_test) is part of the variational form independant from the 'uh' solution
```

$f = \frac{2x(x - 2)}{1 + 30(x - 1)^2}$

```python
#¬†Volumic external forces
material_hardness_test_function = fenics.Expression(('0', '0.5/0.25*(x[0]*(x[0]-2)) / (1 + 30*(x[0]-1)*(x[0]-1))'), degree=2)

body_forces = fenics.project(material_hardness_test_function, V)
```

```python
import matplotlib.pyplot as plt

#fenics.plot(mesh)
fenics.plot(body_forces, title='Loading force in Volume')

plt.xlim(0, 2.)
plt.ylim(0., 1.)
plt.show()
```

```python
Lb = fenics.dot(body_forces, v_test) * fenics.dx
```

```python
# traction?
```

```python
traction = fenics.Constant((0.0, 0.0))
Lt = fenics.dot(traction, v_test) * ds
```

```python
L = Lb + Lt
```

### Define Dirichlet boundary conditions

```python
bc_bottom = fenics.DirichletBC(V, fenics.Constant((0., 0.)), boundaries, 99) #¬†Bottom fixed bc
bc_left = fenics.DirichletBC(V, fenics.Constant((0., 0.)), boundaries, 1) #¬†Left fixed bc
bc_right = fenics.DirichletBC(V, fenics.Constant((0., 0.)), boundaries, 2)  #¬†Right fixed bc

bcs = [bc_bottom] #¬†We choose fixed bottom bc
```

### Solve the linear variational problem

```python
uh = fenics.Function(V) #¬†'uh' is the approximation with FEM/FEniCS of real solution 'u'
fenics.solve(a==L, uh, bcs) # solve linear variational problem

#problem = LinearVariationalProblem(a, L, u, bcs)
#solver = LinearVariationalSolver(problem)
#solver.parameters["symmetric"] = True
#solver.solve()
```

### Visualization

#### Displacement

```python
import matplotlib.pyplot as plt
#fenics.plot(uh)
fenics.plot(1e5*uh, title='Displacement', mode='displacement')
plt.show()

#import vedo.dolfin
#vedo.dolfin.plot(uh, text="uh", style='paraview', interactive=True) # extra rendering window
```

#### Cauchy stress

```python
T = fenics.TensorFunctionSpace(mesh, "DG", 1)

stress = fenics.Function(T)
stress.assign(fenics.project(sigma(uh), T))

fenics.plot(stress[0, 0], title='Cauchy stress œÉ[0,0]')
plt.show()

fenics.plot(stress[1, 1], title='Cauchy stress œÉ[1,1]')
plt.show()

fenics.plot(stress[0, 1], title='Cauchy stress œÉ[0,1]') 
plt.show()

fenics.plot(stress[1, 0], title='Cauchy stress œÉ[1,0]')
plt.show()
```
